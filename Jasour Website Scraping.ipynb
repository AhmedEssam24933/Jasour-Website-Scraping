```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Path to your chromedriver.exe
driver_path = r"C:\Users\ahmed\msedgedriver.exe"

# Initialize WebDriver using Service
service = Service(driver_path)
driver = webdriver.Edge(service=service)

try:
    # Open the target URL
    driver.get("https://jasour.net/")
    wait = WebDriverWait(driver, 10)

    # Use ActionChains to scroll and click the "Load More" button
    while True:
        try:
            # Locate the "Load More" button by its class name
            load_more_button = wait.until(
                EC.element_to_be_clickable((By.CLASS_NAME, "loadmore-trigger"))
            )

            # Scroll the page to make the button visible if needed
            actions = ActionChains(driver)
            actions.move_to_element(load_more_button).perform()

            # Click the "Load More" button
            load_more_button.click()
            time.sleep(2)  # Wait for new content to load
        except Exception as e:
            print("No more 'Load More' button found or loading failed:", e)
            break

    # Get the full page source after loading all content
    page_source = driver.page_source
finally:
    driver.quit()

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(page_source, "html.parser")

# Extract <a> tags with class 'p-url'
links = soup.find_all("a", class_="p-url")

# Extract only the href attribute
urls = [link.get("href") for link in links if link.get("href")]

# Create a DataFrame with the URLs
df = pd.DataFrame(urls, columns=["URL"])
```

```python
all_data = []

# Loop through each URL in the DataFrame
for index, row in urls.iterrows():
    url = row['URL']  # Get the URL from the DataFrame
    
    # Fetch the page content
    response = requests.get(url)
    page_source = response.text

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Extract the title from <h1> tag with class 's-title'
    title = soup.find('h1', class_='s-title').get_text(strip=True)

    # Find the div with the class "entry-content rbct clearfix is-highlight-shares"
    content_div = soup.find("div", class_="entry-content rbct clearfix is-highlight-shares")

    # Extract all <p> tags inside this div
    paragraphs = content_div.find_all("p")

    # List to hold the paragraph texts for the current URL
    paragraph_texts = []

    # Append the title first
    paragraph_texts.append(title)

    # Iterate over the paragraphs and append the text if they don't contain <strong>
    for p in paragraphs:
        # Skip <p> tags containing <strong>
        if not p.find('strong'):
            paragraph_texts.append(p.get_text(strip=True))

    # Append the extracted text and URL as a tuple to the all_data list
    for text in paragraph_texts:
        all_data.append([url, text])

# Create a DataFrame from the list of extracted data
new_df = pd.DataFrame(all_data, columns=["URL", "Text"])
```
